# Ditribution Plots

We have the following plots to plot a Distribution
1. distplot
2. jointplot
3. pairplot
4. rugplot
5. kdeplot

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

The curve in the graph is the KDE (Kernel Density Estimation). We can remove with an argument kde = False as follows

```python
sns.distplot(tips['total_bill'], kde = False)
```

We can slso change the number of bins with an arugument like this

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

There is an argument called kind, which basically allows us to plot a the scatter plot quite differently!

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips)
```

We can add a hue argument to the pairplot basically to differentiate the categorical values.

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.pairplot(tips, hue = 'smoker')
```

```python
sns.rugplot(tips['total_bill'])
```

This is very similar to the distplot. Let me plot the distplot again now, so that the content delivered is clear.

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical Plots

We have the following plots to plot the ditribution of Categorical Values

1. barplot
2. countplot
3. boxplot
4. violinplot
5. stripplot
6. swarmplot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

In the bar plot, fix the categorical variable in the X axis and the numerical value in the Y axis

We have an argument called the estimator, which essentially is used to apply any function that we need. By default it is set to average, but we can change to anything that we need, as shown in the below example.

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

Box plots and Violin plots are used to see the distribution of the Categorical variables.

What is a box plot? - https://www.youtube.com/watch?v=mhaGAaL6Abw

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

Usually box plots are used when your audience are the management and executives.

Violin plots give more information and they can be used when the audience are Data Scientists. Violin plots comes with a cost that more time has to be spent for interpretation.

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex', dodge=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips, color='black')
```

# Grids

These are just used to apply any additional transformations if we need in our project.

There are two grids that are popularly used.

1. PairGrid
2. FacetGrid

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

PairGrid gives us a lot more control when compared to pairplot. Lets see how we do that.

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```
